```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("Y:/BTS/BTS_DataScienceFoundations/Session_2/2_titanic_dataset.csv")
df.head()
```

```python
sns.countplot(df['Pclass'])
plt.show()
```

```python
sns.barplot(x="Sex", y="Age", data=df)
plt.show()
```

```python
sns.pointplot(x="Pclass", y="Survived", hue="Sex", data=df, palette={"male":"blue","female":"yellow"})
plt.show ()
```

```python
sns.swarmplot(x="Survived", y="Fare", data=df)
plt.show()
```

```python
sns.factorplot("Survived","Age","Parch",data=df, kind="bar", palette="muted",legend=True, size=4,aspect=2)
plt.show()
```

```python
g=sns.FacetGrid(df,col="Pclass", row="Survived", margin_titles=True)
g.map(plt.scatter,"Age","SibSp")

plt.show()
```
